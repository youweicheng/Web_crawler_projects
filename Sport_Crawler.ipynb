## acquire title and url from sport web

```python
import requests
from lxml import etree
import json

class SportCrawler(object):
    def __init__(self):
        
        self.url = "https://www.playsport.cc/forum.php?pageno="
        
        self.headers = {
            "User-Agent": 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'
        }
        
        self.data_list = []
    # 1. send request   
    def get_response(self, url):
        # tell whether it is gbk or utf-8, head -- meta -- charset=""
        # if gbk, decode with gbk
        data = requests.get(url=url, headers=self.headers).content.decode('utf-8')
        
        return data
    # 2. parse data
    def parse_data(self, data):
        # use xpath to parse data, and get title and url
        # a. transform to xpath data
        xpath_data = etree.HTML(data)
        # b. use xpath method 1. handmade 2. copy Xpath, but it need to modify
        # //*[@id="forumTable"]/tbody/tr[48]/td[1]/span/a
        title_list = xpath_data.xpath('//td/span/a/text()')
        
        url_list = xpath_data.xpath('//td/span/a/@href')
        
        for index, title in enumerate(title_list):
            
            data_dict = dict()
            
            data_dict['title'] = title
            
            data_dict['url'] = "https://www.playsport.cc/" + url_list[index]
            
            self.data_list.append(data_dict)
            
            print(data_dict)
            
    # 3. save data  
    def save_data(self):
        # list to str
        # write() argument must be str, not list
        data_str = json.dumps(self.data_list)
        
        with open("sport.json", "w") as f:
            f.write(data_str)
    # 4. run it        
    def run(self):
        
        for i in range(1,2):
            # a. complete the url
            url = self.url + str(i)
            # b. send request
            data = self.get_response(url)
            # c. parse data
            self.parse_data(data)
        # d. save data   
        self.save_data()
        
        
SportCrawler().run()
            
            
            
            
        
        
```
